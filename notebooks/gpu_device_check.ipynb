```python
import numpy as np
import torch

# Check for CUDA availability
print(f"PyTorch version: {torch.__version__}")
print(f"CUDA available: {torch.cuda.is_available()}")
```

```python
# Determine which device to use
if torch.cuda.is_available():
    device = torch.device("cuda")
    print("✓ GPU is available!")
    print(f"  Using device: {device}")
else:
    device = torch.device("cpu")
    print("✗ GPU not available, using CPU instead")
    print(f"  Using device: {device}")

print(f"\nDevice type: {device.type}")
if device.type == "cuda":
    print(f"Device index: {device.index}")
```

```python
# Display detailed GPU information if available
if torch.cuda.is_available():
    print("GPU Device Details:")
    print("-" * 50)
    print(f"Number of GPUs: {torch.cuda.device_count()}")

    for i in range(torch.cuda.device_count()):
        print(f"\nGPU {i}:")
        print(f"  Name: {torch.cuda.get_device_name(i)}")
        print(f"  Capability: {torch.cuda.get_device_capability(i)}")

        # Memory information
        gpu_memory = torch.cuda.get_device_properties(i).total_memory / (1024**3)
        print(f"  Total Memory: {gpu_memory:.2f} GB")

        # Current memory usage
        allocated = torch.cuda.memory_allocated(i) / (1024**3)
        reserved = torch.cuda.memory_reserved(i) / (1024**3)
        print(f"  Allocated Memory: {allocated:.2f} GB")
        print(f"  Reserved Memory: {reserved:.2f} GB")
else:
    print("No GPU available - CPU information:")
    print("-" * 50)
    print(f"CPU Count: {np.cpu_count()}")
```

```python
# Run a simple test computation on the selected device
print("Running test computation...")
print("-" * 50)

try:
    # Create a tensor on the device
    test_tensor = torch.randn(1000, 1000, device=device)

    # Perform a simple matrix multiplication
    result = torch.matmul(test_tensor, test_tensor)

    # Verify the result
    if result.device == device:
        print(f"✓ Successfully ran computation on {device}")
        print(f"  Tensor shape: {result.shape}")
        print(f"  Tensor device: {result.device}")
        print(
            "\n✓ GPU is working correctly!"
            if device.type == "cuda"
            else "\n✓ CPU is working correctly!"
        )
    else:
        print(f"✗ Computation moved to unexpected device: {result.device}")

except Exception as e:
    print(f"✗ Error during computation: {e}")
```
