```python
import itertools

import numpy as np
```

```python
# Generate synthetic data
N = 5
np.random.seed(37)
X = np.random.randint(1, 7, size=(N, 5))
X_petals = np.where(X == 3, 2, 0) + np.where(X == 5, 4, 0)
Y = X_petals.sum(axis=1)
X.shape, Y.shape
```

```python
# Hypoethesis space
H = np.array(list(itertools.product(range(0, 10), repeat=6)))
H.shape
```

```python
# Predictions for each hypothesis
W = H[:, X - 1].sum(axis=2)
W.shape
```

```python
# Each hypothesis is either ruled out (False) if there
# is a counterexample in W, or possible (True) if all
# data are consistent with it.
P = np.array([np.all(W[i] == Y) for i in range(len(W))])
P.shape
```

```python
# number of admissible hypotheses
n_possible = P.sum()

print("Remaining Possible hypotheses:", n_possible)

# print all  admissible hypotheses
if n_possible == 0:
    print("No admissible hypotheses found.")
elif n_possible == 1:
    print("Unique Solution:", H[np.where(P)[0]][0])
else:
    print("Possible Solutions:", H[np.where(P)[0]])
```
